```python
import matplotlib.pyplot as plt
%matplotlib inline
import math
import numpy as np

left = []
lf = []

import wave
wr = wave.open("non.wav")
da = np.frombuffer(wr.readframes(44100), dtype=np.int16)
left = da[0::2]

#volume = 100.0
#bar = 32767 * volume / 100.0
#for i in range(44100):
#    a = bar*math.sin(2 * math.pi * 1000 * i * 1 / 44100)
#    b = bar*math.sin(2 * math.pi * 100 * i * 1 / 44100)
#    left.append(a)
    
    
lf = np.fft.rfft(left)
plt.plot(left)
```

```python
plt.figure(1)
a = plt.subplot(211)
r = 2**16/2
a.set_ylim([-r, r])
a.set_xlabel('time [s]')
a.set_ylabel('sample value [-]')
#x = np.arange(len(left))/len(left)
print len(left)
plt.plot(left)
b = plt.subplot(212)
b.set_xscale('log')
b.set_xlabel('frequency [Hz]')
b.set_ylabel('|amplitude|')
plt.plot(abs(lf))
```
